```python
import numpy as np
import matplotlib.pyplot as plt
from pycbc.waveform import get_td_waveform
from pycbc.detector import Detector

%matplotlib inline
```

```python
apx = 'SEOBNRv4' # waveform approximant

# generate plus and cross polarization of a waveform
hp, hc = get_td_waveform(approximant=apx,
                         mass1=10,
                         mass2=10,
                         spin1z=0.9,
                         spin2z=0.4,
                         inclination=1.23,
                         coa_phase=2.45,
                         delta_t=1.0/4096,
                         f_lower=40)
```

```python
det_h1 = Detector("H1") # load the detector
```

```python
end_time = 1192529720 # GPS time -- What is this????
declination = 0 #0.65
right_ascension = 0 #4.67
polarization = 0 #2.34
hp.start_time += end_time
hc.start_time += end_time
```

```python
signal_h1 = det_h1.project_wave(hp, hc,  right_ascension, declination, polarization)
```

```python
plt.plot(signal_h1.sample_times, signal_h1, label='H1')
```
